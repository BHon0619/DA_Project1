Loading Data

```python
import pandas as pd
df = pd.read_csv('titanic3.csv')
print(df.head())
```

In order to better describe the null value in the boat column, I am changing the column name to lifeboat and changing the null values to DNB standing for Did Not Board.

```python
#change column name
df.rename(columns={'boat':'lifeboat'}, inplace=True)

#replace NaN values in boat colum to DNB
df['lifeboat'] = df['lifeboat'].fillna('DNB')
df.to_csv('updated_titanic3.csv', index=False)
print(df.head())

    
```

Convert Fare to currency format

```python

df_new1 = pd.read_csv('update1_titanic3.csv')
# Convert 'fare' column to currency format
df_new1['fare'] = df_new1['fare'].apply(lambda x: f"${x:,.2f}")
#load updated file
df_new1.to_csv('update1_titanic3.csv')
print(df_new1.head())
```

Convert decimal age to years and months

```python

df_new1= pd.read_csv('update1_titanic3.csv')

# Function to convert decimal age to years and months, handling NaN values
def convert_age(age):
    if pd.isna(age):  # Check if the value is NaN
        return "Unknown"  # Or you can return NaN itself if desired
    years = int(age)  # Get the integer part for years
    months = round((age - years) * 12)  # Get the decimal part and convert to months
    return f"{years} years {months} months"

# Apply the function to the 'age' column
df_new1['age_in_years_months'] = df['age'].apply(convert_age)

# Save or display the updated DataFrame
df_new1.to_csv('updated_titanic3_with_age.csv', index=False)

# Print the updated DataFrame to check the new column
print(df_new1[['age', 'age_in_years_months']].head())
print (df_new1.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'sex' column contains 'male'/'female' and 'survived' contains 1 for survived, 0 for not survived
plt.figure(figsize=(8, 6))

# Create a bar plot to show the survival rates for each gender
sns.countplot(x='sex', hue='survived', data=df_new1, palette='Set1')

# Adding labels and title
plt.title('Survival Counts by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'pclass' contains passenger classes (1, 2, 3) and 'survived' contains 1 for survived, 0 for not survived
plt.figure(figsize=(8, 6))

# Create a count plot to show survival counts for each passenger class
sns.countplot(x='pclass', hue='survived', data=df_new1, palette='Set2')

# Adding labels and title
plt.title('Survival Counts by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

```python
# Scatter plot for age vs fare, colored by survival status
plt.figure(figsize=(8, 6))

sns.scatterplot(x='age', y='fare', data=df_new1, hue='survived', palette='Set1', style='survived', markers=['o', 'X'])

# Adding labels and title
plt.title('Scatter Plot of Age vs Fare with Survival Status', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])

# Show the plot
plt.show()
```
